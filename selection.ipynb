# Natural Selection
### Genotype Viability

We have thus far considered evolution in the absence of selection, studying the effects of genetic drift on the distribution of allele frequencies over time. Now we would like to think about natural selection, and how allele frequencies respond when genotypes differ in their fitness in different environments. To begin, we will consider the simplest model from our earlier investigations: An infinitely large hermaphroditic population, mating at random in non-overlapping generations. This is the Hardy-Weinberg model.

First, we will refresh our memory with Hardy-Weinberg equilibrium in the absence of selection:

$$p^2 + 2pq + q^2 = 1$$

If each genotype is more or less fit, then its likelihood of passing its alleles onto the next generation will be multiplied by some constant, $w_{ij}$, as follows:

$$\omega_{11}p^2 + 2\omega_{12}pq + \omega_{22}q^2 = \bar{\omega}$$

where $\bar{\omega}$ is known as the mean fitness of the population, and chosen such that:

$$\frac{\omega_{11}p^2 + 2\omega_{12}pq + \omega_{22}q^2}{\bar{\omega}} = 1$$

We can now start to ask questions about how genotype frequencies will change over time. In the next generation, we know that 

$$
p' = \frac{\omega_{11}p^2 + \omega_{12}pq}{\bar{\omega}} \\
\Delta_{s}p = \frac{\omega_{11}p^2 + \omega_{12}pq}{\bar{\omega}} - p
$$

This can be rearranged to give the rather clunky-looking formula below. Whilst ugly, this is nonetheless a hugely important equation, and one which tells us very much about the behaviour of allele frequency distributions in the presence of natural selection.

$$
\Delta_{s}p' = \frac{pq[p(\omega_{11} - \omega_{12}) + q(\omega_{12} - \omega_{22})]}{\omega_{11}p^2 + 2\omega_{12}pq + \omega_{22}q^2}\label{}
$$

### Relative fitness and different kinds of selection

Absolute genotype viabilities are helpful in theory, but in practice it is difficult to assign values to each genotype. Much more useful is to measure relative fitness. If we divide each term in the above equation by $\omega_{11}$ then nothing changes, but it is significantly simpler to work with. A common and usful substituion is $\frac{\omega_{12}}{\omega_{11}} = 1 - hs$ and $\frac{\omega_{22}}{\omega_{11}} = 1 - s$. This simplifies to

$$
\Delta_{s}p = \frac{pqs[ph + q(1 - h)]}{1 - 2pqhs - q^2s}
$$

The parameter $s$ is the selection coeffiecient, and describes how quickly the allele frequency distribution will reach an equilibrium state. The parameter $h$ is known as the heterozygous effect, and is essentially a measure of the degree of dominance, ranging from overdominance to underdominance, via incomplete dominance.

$$
\begin{aligned}
&h < 0 &overdominance\\
&h = 0 &A1~dominant\\
&0 < h < 1 &incomplete~dominance\\
&h = 1 &A1~recessive\\
&h > 1 &underdominance
\end{aligned}
$$


```python
%matplotlib inline
from ipywidgets import interact, interactive
from selection import plot_dominance
```

```python
w = interactive(plot_dominance, het=(-1., 2.), sel=(0., 1.))
display(w)
```

So we see that there is a fairly complex relationship between the selection coefficient, the heterozygous effect and the rate of change of $\Delta_{s}p$. For $0 < h < 1$, the allele is incompletely dominant, with the fitness of heterozygotes lying somewhere between that of the two homozygous genotypes. Alternatively, an allele may be overdominant, in which case balancing selection results - an equlibrium value such that $p < 1$ will eventually be reached, with the speed at which it gets there being dependent on the strength of selection. The classic example of this is sickle-cell haemoglobin. Here, the mutant allele is maintained in the population because the heterozygous genotype affords carriers some resistance to malaria, which outweighs the negative effects of mild cell sickling.

```python
x = plot_dominance(het=-0.5, sel=0.2)
```

In contrast, disruptive selection is very rare, since the equilibrium is an unstable one, depending on the initial frequency of p, the frequency will either always increase towards 1 or decrease towards 0. In practice, demonstrating disruptive selection is likely to be extremely difficult.
